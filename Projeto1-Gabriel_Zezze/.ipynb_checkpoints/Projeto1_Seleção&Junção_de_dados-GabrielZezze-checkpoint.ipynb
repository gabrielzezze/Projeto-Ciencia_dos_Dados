```python
#Imports#
%matplotlib inline
import pandas as pd
import numpy as np
import math as math
from scipy import stats
import matplotlib.pyplot as plt
```

```python
#Coletar dados do world bank e tranferir para python#
IDH = pd.read_excel('hdi_index.xlsx') #Indice de IDH#

NetUsers = pd.read_excel('internet_users.xlsx') #Indice de Usuarios de internet#

Corrup = pd.read_excel('corruption_perception.xlsx') #Indice de corrupção#
```

```python
#Selecionando dados do ano escolhido para analise(2012)#
Lista = ['geo',2012]
#IDH#
IDH_2012 = IDH.loc[:,Lista]
IDH2012 = IDH_2012.rename(index=str, columns={"geo":"Países"})
#Usuários de Internet#
NetUsers_2012 = NetUsers.loc[:,Lista]
NetUsers2012 = NetUsers_2012.rename(index=str, columns={"geo":"Países"})

#Corrupção#
Corrup_2012 = Corrup.loc[:,Lista]
Corrup2012 = Corrup_2012.rename(index=str, columns={"geo":"Países",2012:"2012_Corrupção"})
```

```python
#Alterando Indices para Juntar dados em um mesmo DataFrame#
IDH2012_idx = IDH2012.set_index('Países')

NetUsers2012_idx = NetUsers2012.set_index('Países')

Corrup2012_idx = Corrup2012.set_index('Países')
```

```python
#Juntando dados em um mesmo dataFrame#
df = IDH2012_idx.join(NetUsers2012_idx, how='inner', lsuffix='_IDH', rsuffix='_Usuários_Internet')

Data = df.join(Corrup2012_idx,how='inner')
```

```python
#Escrevendo dados em uma planilha no Excel#
w = pd.ExcelWriter('Projeto1_GabrielZezze.xlsx',engine='xlsxwriter')
Data.to_excel(w,sheet_name='Sheet1')
w.save()
#Escrevendo dados em arquivo CSV#
Data.to_csv('Projeto1_GabrielZezze(CSV).csv')
```
